# **CREAMOS LAS TABLAS**

```python
#1.IMPORTAMOS LAS LIBRERÍAS NECESARIAS:

#para poder hacer queries en Python
import sqlite3

#para visualizar la salida
import pandas as pd



#---------------------------------------------------------

# 2. NOS CONECTAMOS A LA BASE DE DATOS DE sqlite3

conn = sqlite3.connect('recetas.db')
print("Opened database successfully");




#---------------------------------------------------------

# 3. Y AHORA YA SÍ, CREAMOS LAS TABLAS DEL PROYECTO:


#LA TABLA RECETAS:
conn.execute('''
CREATE TABLE IF NOT EXISTS recetas(
    receta_id INT PRIMARY KEY,
    nombre_receta VARCHAR(255) NOT NULL,
    categoria VARCHAR(255) NOT NULL,
    tiempo_preparacion INT NOT NULL,
    dificultad VARCHAR(255) NOT NULL
);''')
conn.commit()


#LA TABLA INGREDIENTES:
conn.execute('''
CREATE TABLE IF NOT EXISTS ingredientes(
    ingrediente_id INT PRIMARY KEY,
    nombre_ingrediente VARCHAR(255) NOT NULL,
    cantidad_disponible INT NOT NULL,
    unidad_medida VARCHAR(255) NOT NULL
);''')
conn.commit()



#LA TABLA RECETAS_INGREDIENTES:
conn.execute('''
CREATE TABLE IF NOT EXISTS recetas_ingredientes(
    receta_id INT,
    ingrediente_id INT,
    cantidad INT,
    PRIMARY KEY (receta_id, ingrediente_id),
    FOREIGN KEY (receta_id) REFERENCES recetas (receta_id),
    FOREIGN KEY (ingrediente_id) REFERENCES ingredientes (ingrediente_id)
);''')
conn.commit()
```

# **CONSULTAS SQL**

# 1. Mostrar todas las recetas disponibles.

```python
#Creamos la query y la guardamos en la variable qry:
qry = '''
SELECT nombre_receta
FROM recetas;
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```

# 2. Mostrar todas los ingredientes necesarios para una receta específica. Por ejemplo para hacer hummus.

```python
#Esto lo he hecho con INNER JOIN
#Creamos la query y la guardamos en la variable qry:
qry = '''
SELECT ingredientes.nombre_ingrediente, recetas_ingredientes.cantidad, ingredientes.unidad_medida
FROM recetas
INNER JOIN recetas_ingredientes ON recetas.receta_id = recetas_ingredientes.receta_id
INNER JOIN ingredientes ON recetas_ingredientes.ingrediente_id = ingredientes.ingrediente_id
WHERE recetas.nombre_receta = 'hummus';
'''


#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head(10)
```

# 3. Mostrar las receta que se pueden preparar con los ingredientes disponibles en la despensa de Iván.

```python
#LO HE HECHO CON: JOINs + GROUP BY + HAVING MIN

#Creamos la query y la guardamos en la variable qry:
qry = '''
SELECT recetas.nombre_receta
FROM recetas
JOIN recetas_ingredientes ON recetas.receta_id = recetas_ingredientes.receta_id
JOIN ingredientes ON recetas_ingredientes.ingrediente_id = ingredientes.ingrediente_id
GROUP BY recetas.receta_id, recetas.nombre_receta
HAVING MIN(ingredientes.cantidad_disponible >= recetas_ingredientes.cantidad);
'''
#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head(10)
```

# 4. Mostrar las recetas que requieren menos de 2 litros de aceite de oliva.

```python
#LO HE HECHO CON: JOINs + GROUP BY + HAVING MIN

#Creamos la query y la guardamos en la variable qry:
qry = '''
SELECT recetas.nombre_receta, recetas_ingredientes.cantidad, ingredientes.unidad_medida
FROM recetas
JOIN recetas_ingredientes ON recetas.receta_id = recetas_ingredientes.receta_id
JOIN ingredientes ON recetas_ingredientes.ingrediente_id = ingredientes.ingrediente_id
WHERE ingredientes.nombre_ingrediente = 'aceite de oliva' AND recetas_ingredientes.cantidad < 2;
'''
#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head(10)
```

# **3. CONSULTAS SQL:**

# 3.1 ¿Cuántas pizzas de cada tipo se han vendido a partir del día 2024-02-14?

```python
qry = '''
SELECT pizza_types.pizza_name, COUNT(orders.pizza_id) AS cantidad_vendida
FROM orders
JOIN pizza_types ON orders.pizza_id = pizza_types.pizza_id
WHERE orders.order_time > '2024-02-14'
GROUP BY pizza_types.pizza_name;
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```

# 3.2. ¿Cuál es el repartidor con mejor calificación promedio?

```python
qry = '''
SELECT name, rating
FROM runners
ORDER BY rating DESC
LIMIT 1;
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```

# 3.3. ¿Cuál es el tiempo promedio de entrega de pedidos?

```python
qry = '''
SELECT AVG((strftime('%s', delivery_time) - strftime('%s', order_time)) / 60.0) AS tiempo_promedio_entrega_minutos
FROM orders;
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```

# 3.4. ¿Cuál es el tipo de pizza más popular entre los clientes?

```python
qry = '''
SELECT pizza_types.pizza_name, COUNT(orders.pizza_id) AS cantidad_vendida
FROM orders
JOIN pizza_types ON orders.pizza_id = pizza_types.pizza_id
GROUP BY pizza_types.pizza_name
ORDER BY cantidad_vendida DESC
LIMIT 1;
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```

**Descripción:**

Nod-Foody es una startup que ofrece un servicio de entrega de alimentos a través de una plataforma en línea.
Para mejorar su eficiencia y experiencia del usuario, necesitan una base de datos que les permita gestionar los restaurantes registrados, los usuarios y los pedidos realizados en su plataforma.

**Datos:**

La base de datos de Nod-Foody contiene tres tablas principales:
- restaurants (restaurantes),
- users (usuarios)
- orders (pedidos).

Aquí está la estructura de cada tabla:

**Tabla restaurants:**

- restaurant_id (int, clave primaria): Identificador único del restaurante.
- name (string): Nombre del restaurante.
- location (string): Ubicación del restaurante.
- cuisine (string): Tipo de cocina ofrecida por el restaurante (por ejemplo, italiana, mexicana, asiática).

**Tabla users:**

- user_id (int, clave primaria): Identificador único del usuario.
- name (string): Nombre del usuario.
- email (string): Dirección de correo electrónico del usuario.
- address (string): Dirección de entrega del usuario.

**Tabla orders:**

- order_id (int, clave primaria): Identificador único del pedido.
- user_id (int): Identificador único del usuario que realizó el pedido.
- restaurant_id (int): Identificador único del restaurante del cual se realizó el pedido.
- order_time (timestamp): Fecha y hora en la que se realizó el pedido.
- delivery_time (timestamp): Fecha y hora en la que se entregó el pedido al usuario.
- total_amount (decimal): Monto total del pedido.

Proporciona el script SQL que incluya la creación de las tablas, la inserción de datos y las consultas realizadas para cumplir con las tareas mencionadas.

```python
import sqlite3

import pandas as pd

# Conexión a la base de datos SQLite
conn = sqlite3.connect('Nod-Foody.db')
cursor = conn.cursor()
```

# **1. CREAMOS LAS TABLAS**

```python
# TABLA RESTAURANTS
cursor.execute('''
CREATE TABLE IF NOT EXISTS restaurants (
    restaurant_id INTEGER PRIMARY KEY,
    name TEXT,
    location TEXT,
    cuisine TEXT
);
''')

# TABLA USERS
cursor.execute('''
CREATE TABLE IF NOT EXISTS users (
    user_id INTEGER PRIMARY KEY,
    name TEXT,
    email TEXT,
    address TEXT
);
''')

# TABLA ORDERS
cursor.execute('''
CREATE TABLE IF NOT EXISTS orders (
    order_id INTEGER PRIMARY KEY,
    user_id INTEGER,
    restaurant_id INTEGER,
    order_time TIMESTAMP,
    delivery_time TIMESTAMP,
    total_amount REAL,
    FOREIGN KEY (user_id) REFERENCES users(user_id),
    FOREIGN KEY (restaurant_id) REFERENCES restaurants(restaurant_id)
);
''')
```

# **3. Consultas SQL para responder las siguientes preguntas:**

# 3.1. ¿Cuántos pedidos se han realizado en cada restaurante durante el último mes?

```python
qry = '''
SELECT restaurants.name AS restaurant_name, COUNT(orders.order_id) AS total_pedidos
FROM orders
JOIN restaurants ON orders.restaurant_id = restaurants.restaurant_id
WHERE orders.order_time >= DATE('now', '-1 month')
GROUP BY restaurants.name;
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```

# 3.2. ¿Cuál es el usuario que ha realizado más pedidos?


```python
qry = '''
SELECT users.name AS usuario, COUNT(orders.order_id) AS total_pedidos
FROM orders
JOIN users ON orders.user_id = users.user_id
WHERE orders.order_time >= DATE('now', '-1 month')
GROUP BY users.name
ORDER BY total_pedidos DESC
LIMIT 1;
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```

# 3.3. ¿Cuál es el restaurante más popular entre los usuarios?

```python
qry = '''
SELECT restaurants.name AS restaurante_mas_popular, COUNT(orders.order_id) AS total_pedidos
FROM orders
JOIN restaurants ON orders.restaurant_id = restaurants.restaurant_id
GROUP BY restaurants.name
ORDER BY total_pedidos DESC
LIMIT 1;
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```

# 3.4. ¿Cuál es el monto total de ventas realizadas en la plataforma durante el último año?

```python
qry = '''
SELECT SUM(total_amount) AS total_ventas
FROM orders
WHERE order_time >= DATE('now', '-1 year');
'''

#para visualizar la salida, usaremos pandas:
df = pd.read_sql_query(qry, conn)
df.head()
```
